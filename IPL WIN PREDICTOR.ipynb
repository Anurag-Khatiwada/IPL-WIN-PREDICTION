```python
import numpy as np
import pandas as pd
```

```python
match=pd.read_csv('IPL_Matches_2008_2022 (1).csv')
deliveries=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
```

```python
match.head()
```

```python
deliveries.head()
```

```python
total_score_df=deliveries.groupby(['ID','innings']).sum()['total_run'].reset_index()
```

```python
total_score_df
```

```python
total_score_df=total_score_df[total_score_df['innings']==1]
```

```python
match_df=match.merge(total_score_df[['ID','total_run']],left_on='ID',right_on='ID')
```

```python
match_df.shape
```

```python
match_df['Team1'].unique()
```

```python
teams=[
    'Rajasthan Royals', 
    'Royal Challengers Bangalore',
    'Sunrisers Hyderabad', 
    'Delhi Capitals', 
    'Chennai Super Kings',
    'Gujarat Titans', 
    'Lucknow Super Giants', 
    'Kolkata Knight Riders',
    'Mumbai Indians',
    'Kings XI Punjab',
]
```

```python
match_df['Team1']=match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2']=match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1']=match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2']=match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['Team1']=match_df['Team1'].str.replace('Punjab Kings','Kings XI Punjab')
match_df['Team2']=match_df['Team2'].str.replace('Punjab Kings','Kings XI Punjab')
```

```python
match_df=match_df[match_df['Team1'].isin(teams)]
match_df=match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
match_df=match_df[['ID','City','WinningTeam','total_run','Team1','Team2']]
```

```python
delivery_df=match_df.merge(deliveries,on='ID')
```

```python
second_inning=delivery_df[delivery_df['innings']==2]
```

```python
second_inning.shape

```

```python
#finction to determine boaling team
def determine_boling_team(row):
    if row['BattingTeam']==row['Team1']:
        return row['Team2']
    else:
        return row['Team1']

second_inning['BoalingTeam']=second_inning.apply(determine_boling_team, axis=1)
```

```python
second_inning.head()
```

```python
print(second_inning['total_run_y'].dtype)
```

```python
# second_inning['total_run_y'] = pd.to_numeric(second_inning['total_run_y'], errors='coerce')

second_inning['current_score']=second_inning.groupby('ID')['total_run_y'].cumsum()
```

```python
second_inning['runs_left']=second_inning['total_run_x']-second_inning['current_score']
```

```python
second_inning
```

```python
second_inning['balls_left']=120-(second_inning['overs']*6+second_inning['ballnumber'])
```

```python
second_inning.tail()
```

```python
second_inning['player_out']=second_inning['player_out'].fillna('0')
second_inning['player_out']=second_inning['player_out'].apply(lambda x:x if x=='0' else '1')
second_inning['player_out']=second_inning['player_out'].astype('int')
wickets=second_inning.groupby('ID')['player_out'].cumsum().values
second_inning['wickets_left']=10-wickets
```

```python
second_inning['crr']=second_inning['current_score']/((120-second_inning['balls_left'])/6)
```

```python
second_inning['rrr']=(second_inning['runs_left']*6)/second_inning["balls_left"]
```

```python
def result(row):
    if row['BattingTeam']==row["WinningTeam"]:
        return 1
    else:
        return 0

second_inning['result']=second_inning.apply(result,axis=1)
```

```python
second_inning.head()
```

```python
final_df=second_inning[['BattingTeam','BoalingTeam','City','runs_left','balls_left','wickets_left','total_run_x','crr','rrr','result']]
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

```python
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf=ColumnTransformer([
    ('trf',OneHotEncoder(drop='first'),['BattingTeam','BoalingTeam','City'])
]
    ,remainder='passthrough')
    
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
